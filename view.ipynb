```python
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

```python
path = 'D:/Code/data/'
ParaFolder = 'TestPLETH'

TestDataSet1 = np.load(path+ParaFolder+'/TestSet1.npy',allow_pickle = True)
len(TestDataSet1)
```

```python
randomID=[]                        

for i in range(10):
    a = rd.randint(1,len(TestDataSet1)-1)       
    while a in randomID:              
        a = rd.randint(1,len(TestDataSet1)-1)
    randomID.append(a) 

plt.figure(figsize=(20,36))

for i in range(10):
    plt.subplot(10,1,i+1)
    plt.axhline(40, xmin=0, xmax=6000, color='lightgray', linestyle='--')
    plt.plot(TestDataSet1[randomID[i]])

plt.show()
```
